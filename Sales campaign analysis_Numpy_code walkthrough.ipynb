```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import sys
```

```python
# Command to display all the columns of a numpy array
np.set_printoptions(threshold=sys.maxsize)
```

```python
#Let's load the data

import csv
path = 'G:/Data Science ML AI/Data course files/KAG_conversion_data.csv'

with open(path) as f:
    adm = csv.reader(f,delimiter=',')
    adm = list(adm)
    
# Remove the header
adm.remove(adm[0])

# Convert the data into a numpy array and store it in sales_data
sales_data = np.array(adm)
```

1) How many unique ad campaigns (xyz_campaign_id) does this data contain ? And for how many times was each campaign run ?

```python
from collections import Counter
print("The Total number of campaigns are:", len(np.unique(sales_data[:,1])))
print("The unique campains are :", (np.unique(sales_data[:,1])))
print("Count for which each campaign was run:",Counter(sales_data[:,1]))
```

2) What are the age groups that were targeted through these ad campaigns ?

```python
age = sales_data[:,3]

age_groups = np.unique(age)

print("The targeted age groups are :",age_groups)
```

3) What was the average, minimum and maximum amount spent on the ads ?

```python
# collect the amount spent first
amount_spent = sales_data[:,8]

#calculate average, min, max with help of above variable

average_spent = round(amount_spent.astype(float).mean(),2)
minimum_spent = round(amount_spent.astype(float).min(),2)
maximum_spent = round(amount_spent.astype(float).max(),2)

#print the amounts
print("Average amount spent on ads is :",average_spent)
print("Minimum amount spent on ads is :",minimum_spent)
print("Maximum amount spent on ads is :",maximum_spent)

```

4) What is the id of the ad having the maximum number of clicks ?

```python
# What were the maximum number of clicks
clicks = sales_data[:,7].astype(int).max()
print("The maximum number of clicks were :",clicks)

# Which was the ad having the maximum number of clicks
max_clicks_ad = sales_data[:,0][sales_data[:,7].astype(int) == clicks]
print("The ad with maximum number of clicks is :",max_clicks_ad)
```

5) How many people bought the product after seeing the ad with most clicks ? Is that the maximum number of purchases in this dataset ?

```python
max_sales = int(sales_data[:,-1][sales_data[:,0] == max_clicks_ad])
print("Number of people who bought the product having maximum ad clicks is: ",max_sales)

# Max value of the Approved_Conversion column
max_purchases = sales_data[:,-1].astype(int).max()

if (max_sales >= max_purchases):
    print("The maximum sales were on this product")
elif (max_sales <= max_purchases):
    print("The maximum number of purchases was",max_purchases)
```

6) So the ad with the most clicks didn't fetch the maximum number of purchases. Let's find the details of the product having maximum number of purchases?

```python
max_purchases1 = sales_data[:,0][sales_data[:,-1].astype(int) == sales_data[:,-1].astype(int).max()]
print("The ad id with maximum number of purchase is :",max_purchases1)

print("The record for this product is as below")
sales_data[sales_data[:,-1].astype(int) == sales_data[:,-1].astype(int).max()]
```

Create a new feature Click Through Rate (CTR) and then concatenate it to the original numpy array?

```python
# Create a new numpy array to calculate CTR

CTR = np.array((sales_data[:,7].astype(float)/sales_data[:,6].astype(float))*100)

# Note the shapes of CTR and the original array are different
print(sales_data.shape)
print("The original shape of CTR is :",CTR.shape)

# Convert CTR to the same shape as that of the original array
CTR = CTR.reshape(-1,1)
print("New shape of CTR is :", CTR.shape)

# Concatenate CTR to the original array
sales_data = np.concatenate((sales_data, CTR),axis = 1)
print(sales_data)
```
